```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#filling all missing values with 9999
data.fillna(999999, inplace=True) 
```

```python
#checking for null values 
null=data.isnull().sum()
null
```

```python
#which data related field has the most postings

# Getting the value counts and plot
plt.figure(figsize=(12, 8))
counts = data['job_simpl'].value_counts()
bars = counts.plot(kind='bar', color=sns.color_palette("pastel"))

# Adding plot labels
plt.title('Number of Data-Related Job Postings by Field', fontsize=18)
plt.xlabel('Data Fields', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation=45)

# Adding count labels on top of bars
for i, v in enumerate(counts):
    bars.text(i, v+5, str(v), horizontalalignment='center', fontsize=12, color='black')

plt.tight_layout()
plt.show()

```

```python
#what are the skills necessary for each of the data related postings this column had no missing vales so the data distributionis accurate 

#Create a figure with 4 subplots
fig, axs = plt.subplots(2, 2, figsize=(18, 14))

# Data Scientist
filtered_data_scientist = data[data['job_simpl'].str.contains('data scientist', case=False)]
sum_data_scientist = filtered_data_scientist[['python_yn', 'spark_yn', 'azure_yn', 'aws_yn', 'excel_yn', 'machine_learning_yn']].sum()
sum_data_scientist = sum_data_scientist.sort_values(ascending=False)
axs[0, 0].bar(sum_data_scientist.index, sum_data_scientist.values)
axs[0, 0].set_title('Distribution of skills a data scientist should have')
axs[0, 0].set_xlabel('Skills')
axs[0, 0].set_ylabel('Skill count')

# Data Analyst
filtered_data_analyst = data[data['job_simpl'].str.contains('data analyst', case=False)]
sum_data_analyst = filtered_data_analyst[['python_yn', 'spark_yn', 'azure_yn', 'aws_yn', 'excel_yn', 'machine_learning_yn']].sum()
sum_data_analyst = sum_data_analyst.sort_values(ascending=False)
axs[0, 1].bar(sum_data_analyst.index, sum_data_analyst.values)
axs[0, 1].set_title('Distribution of skills a data analyst should have')
axs[0, 1].set_xlabel('Skills')
axs[0, 1].set_ylabel('Skill count')

# Data Engineer
filtered_data_engineer = data[data['job_simpl'].str.contains('data engineer', case=False)]
sum_data_engineer = filtered_data_engineer[['python_yn', 'spark_yn', 'azure_yn', 'aws_yn', 'excel_yn', 'machine_learning_yn']].sum()
sum_data_engineer = sum_data_engineer.sort_values(ascending=False)
axs[1, 0].bar(sum_data_engineer.index, sum_data_engineer.values)
axs[1, 0].set_title('Distribution of skills a data engineer should have')
axs[1, 0].set_xlabel('Skills')
axs[1, 0].set_ylabel('Skill count')

# Machine Learning Engineer
filtered_machine_learning_engineer = data[data['job_simpl'].str.contains('machine learning engineer', case=False)]
sum_machine_learning_engineer = filtered_machine_learning_engineer[['python_yn', 'spark_yn', 'azure_yn', 'aws_yn', 'excel_yn', 'machine_learning_yn']].sum()
sum_machine_learning_engineer = sum_machine_learning_engineer.sort_values(ascending=False)
axs[1, 1].bar(sum_machine_learning_engineer.index, sum_machine_learning_engineer.values)
axs[1, 1].set_title('Distribution of skills a machine learning engineer should have')
axs[1, 1].set_xlabel('Skills')
axs[1, 1].set_ylabel('Skill count')

# Adjust spacing between subplots
fig.tight_layout()

# Display the plot
plt.show()
```

```python
# What is the maximum salary for each related field 
#Data scientist
# find the maximum salary for each role 
max_salary_data_scientist= filtered_data_scientist['salary estimate'].max()
max_salary_data_scientist

#Data Analyst
max_salary_data_analyst = filtered_data_analyst['salary estimate'].max()
max_salary_data_analyst

#Data engineer 
max_salary_data_engineer= filtered_data_engineer['salary estimate'].max()
max_salary_data_engineer

#Machine learnig engineer 
max_salary_machine_learning_engineer= filtered_machine_learning_engineer['salary estimate'].max()
max_salary_machine_learning_engineer

max_salaries = [max_salary_data_scientist, max_salary_data_analyst, max_salary_data_engineer, max_machine_learning_engineer]
roles = ['Data Scientist', 'Data Analyst', 'Data Engineer', 'Machine Learning Engineer']

colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']

plt.figure(figsize=(10, 10))
max_salaries, roles, colors = zip(*sorted(zip(max_salaries, roles, colors), reverse=True))
plt.bar(roles, max_salaries, color=colors)

plt.title('Maximum Salaries for Data-Related Roles')
plt.xlabel('Role')
plt.ylabel('Salary ($)')
plt.ylim((0, max(max_salaries)+10000))

# Adding count labels on top of bars
for i, v in enumerate(max_salaries):
    plt.text(i, v+500, str(v), ha='center', fontsize=12, color='black')

plt.tight_layout()
plt.show()
```
